## Hierarchical clustering

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)  
```

```python
df_norm = norm_func (crime.iloc[:,1:])
df_norm
```

```python
from sklearn.preprocessing import MinMaxScaler 
trans = MinMaxScaler() 
data = pd.DataFrame(trans.fit_transform(crime.iloc[:,1:])) 
data 
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch 
z = linkage(df_norm,method="average",metric="euclidean") 
plt.figure(figsize=(15,5)) 
plt.title("Hierarchical Clustering Dendogram") 
plt.xlabel("Index") 
plt.ylabel("Distance") 
sch.dendrogram(
    z,
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering 
import warnings 
warnings.filterwarnings("ignore") 
h_complete = AgglomerativeClustering(n_clusters=4,linkage="average",affinity="euclidean").fit(df_norm) 
cluster_labels = pd.Series(h_complete.labels_) 
cluster_labels 
crime["clust"] = cluster_labels 
crime  
```

## kmeans clustering

```python
from sklearn.cluster import KMeans 
```

```python
X= np.random.uniform(0,1,1000) 
Y = np.random.uniform(0,1,1000) 
X 
```

```python
df_xy = pd.DataFrame(columns=["X","Y"]) 
df_xy
df_xy.X=X
df_xy.Y=Y  
df_xy 
df_xy.plot(x="X",y="Y",kind = "scatter") 
model1 = KMeans(n_clusters=5).fit(df_xy)  
```

```python
df_xy.plot(x ="X",y="Y",c = model1.labels_,kind = "scatter",s = 50,cmap = plt.cm.coolwarm)
```

```python
def norm_func(i): 
        x = (i-i.mean())/(i.max()-i.min()) 
        return (x) 
```

```python
df_norm = norm_func(crime.iloc[:,1:]) 
df_norm.head()
```

```python
fig = plt.figure(figsize = (10,8))
WCSS = [] 
for i in range(1,11): 
    clf = KMeans(n_clusters= i) 
    clf.fit(df_norm) 
    WCSS.append(clf.inertia_) 
plt.plot(range(1,11),WCSS)
plt.title("The Elbow Method") 
plt.ylabel("WCSS") 
plt.xlabel("Number of clusters") 
plt.show()   
```

```python
clf = KMeans(n_clusters= 4) 
```

```python
y_KMeans = clf.fit_predict(df_norm) 
```

```python
y_KMeans 
clf.labels_ 
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler 
```

```python
crime.info()
```

```python
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
x 
```

```python
dbscan = DBSCAN(eps = 2,min_samples=5) 
dbscan.fit(x) 
```

```python
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_,columns=["cluster"]) 
cl
```
